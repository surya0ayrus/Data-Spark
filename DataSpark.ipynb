```python
import pandas as pd
import mysql.connector
from mysql.connector import Error

def connect_mysql():
    """Establish a connection to the MySQL server."""
    try:
        conn = mysql.connector.connect(
            host="localhost",
            user="root",
            password="12345",
            database="globalelectronic",
            port=3306
        )
        print("Connected to MySQL server")
        return conn
    except Error as e:
        print(f"Error connecting to MySQL: {e}")
        return None

def create_tables(conn):
    """Create tables in the selected database."""
    try:
        cursor = conn.cursor()
        cursor.execute('''
            CREATE TABLE IF NOT EXISTS Project2 (
                CustomerKey INT,
                Gender VARCHAR(255) NOT NULL,
                Name VARCHAR(255) NOT NULL,
                City VARCHAR(255) NOT NULL,
                `State Code` VARCHAR(255) NOT NULL,
                State VARCHAR(255) NOT NULL,
                `Zip Code` VARCHAR(255) NOT NULL,  
                Country VARCHAR(255) NOT NULL,
                Continent VARCHAR(255) NOT NULL,
                Birthday DATE,
                `Order Number` INT,
                `Line Item` INT,
                `Order Date` DATE,    
                StoreKey INT,                    
                ProductKey INT,
                Quantity INT,
                `Currency Code` VARCHAR(255) NOT NULL,
                `Square Meters` INT,    
                `Open Date` DATE,
                `Product Name` VARCHAR(255) NOT NULL,
                Brand VARCHAR(255) NOT NULL,
                Color VARCHAR(255) NOT NULL,
                `Unit Cost USD` FLOAT,
                `Unit Price USD` FLOAT,
                SubcategoryKey INT,
                Subcategory VARCHAR(255) NOT NULL,
                CategoryKey INT,
                Category VARCHAR(255) NOT NULL,
                Date DATE,
                Currency VARCHAR(255) NOT NULL,
                Exchange FLOAT,
                Age INT,
                AgeGroup VARCHAR(255) NOT NULL
            )
        ''')
        conn.commit()
        print("Table 'Project2' created successfully or already exists.")
    except Error as e:
        print(f"Error creating table: {e}")

def insert_data_to_mysql(conn, data):
    """Insert data into the table."""
    insert_query = '''INSERT INTO Project2 (
                        CustomerKey, Gender, Name, City, `State Code`, State, `Zip Code`,
                        Country, Continent, Birthday, `Order Number`, `Line Item`, `Order Date`,
                        StoreKey, ProductKey, Quantity, `Currency Code`, `Square Meters`, `Open Date`,
                        `Product Name`, Brand, Color, `Unit Cost USD`, `Unit Price USD`, SubcategoryKey,
                        Subcategory, CategoryKey, Category, Date, Currency, Exchange, Age, AgeGroup)
                      VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s,
                              %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)'''
    try:
        cursor = conn.cursor()
        cursor.executemany(insert_query, data)
        conn.commit()
        print("Values inserted successfully.")
    except Error as e:
        print(f"Error inserting data: {e}")

def load_csv_to_dataframe(csv_file_path):
    """Load CSV file into a DataFrame."""
    try:
        df = pd.read_csv(csv_file_path)
        print("CSV loaded successfully")
        print("Columns in DataFrame:", df.columns)  # Print column names
        return df
    except Exception as e:
        print(f"Error loading CSV file: {e}")
        return None

```

```python
def main():
    """Main function to execute the process."""
    csv_file_path = 'GlobalElectronics_FULL_DETAILS.csv'

    # Load CSV data
    df = load_csv_to_dataframe(csv_file_path)
    if df is not None and not df.empty:
        df.columns = df.columns.str.strip()

        expected_columns = ['CustomerKey', 'Gender', 'Name', 'City', 'State Code', 'State', 'Zip Code',
                            'Country', 'Continent', 'Birthday', 'Order Number', 'Line Item', 'Order Date',
                            'StoreKey', 'ProductKey', 'Quantity', 'Currency Code', 'Square Meters', 'Open Date',
                            'Product Name', 'Brand', 'Color', 'Unit Cost USD', 'Unit Price USD', 'SubcategoryKey',
                            'Subcategory', 'CategoryKey', 'Category', 'Date', 'Currency', 'Exchange', 'Age', 'AgeGroup']

        missing_columns = [col for col in expected_columns if col not in df.columns]
        if missing_columns:
            print(f"Missing columns: {missing_columns}")
            return

        data = df[expected_columns].values.tolist()

        # Connect to MySQL
        conn = connect_mysql()
        if conn:
            try:
                create_tables(conn)
                insert_data_to_mysql(conn, data)
            finally:
                conn.close()
                print("MySQL connection closed.")
        else:
            print("Failed to connect to MySQL.")
    else:
        print("No data to insert.")

if __name__ == "__main__":
    main()
```
